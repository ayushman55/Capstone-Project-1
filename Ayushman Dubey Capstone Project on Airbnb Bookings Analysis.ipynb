# **3. Comparison of reviews with prices.**

```python
price_area = df.groupby(['price'])['number_of_reviews'].max().reset_index()
price_area.head(5)
```

```python
area = price_area['price']
price = price_area['number_of_reviews']

correlation = area.corr(price)
correlation

fig = plt.figure(figsize = (10, 5))

# creating the scatter plot
plt.scatter(area, price)
 
plt.xlabel("Price")
plt.ylabel("Number of Review")
plt.title("Price vs Number of Reviews")
plt.show()

plt.title('Correlation')

# fits the best fitting line to the data
plt.plot(np.poly1d(np.polyfit(area, price, 1))
         (np.unique(price)), color='red')
 
# Labelling axes
plt.xlabel('price')
plt.ylabel('number_of_reviews')

```

##**From comparing reviews with price we can see a trend that most people chooses to stay in place where price is low.**

## **5. Which locations are more populated?**

```python
busiest_local = df.groupby(['host_name','host_id','room_type'])['number_of_reviews'].max().reset_index()
busiest_local = busiest_local.sort_values(by='number_of_reviews', ascending=False).head(10)
busiest_local
```

```python
name = busiest_local['host_name']
reviews = busiest_local['number_of_reviews']

fig = plt.figure(figsize = (10, 5))
 
# creating the stem plot

fig, ax = plt.subplots()

ax.stem(name, reviews)
plt.xlabel("Name of the Owner")
plt.ylabel("Number of Reviews")
plt.title("Busiest Hosts")

plt.show()
```

## **6. Types of Hotels?**

```python
hotel = df.groupby(['neighbourhood_group','room_type'])['minimum_nights'].count().reset_index()
hotel = hotel.sort_values(by='minimum_nights', ascending=False)
hotel
```

```python
room_type = hotel['room_type']
stayed = hotel['minimum_nights']

fig = plt.figure(figsize = (10, 5))
 
# creating the stack plot
 
fig, ax = plt.subplots()

ax.stackplot(room_type, stayed)
plt.title("Traffic Areas")
plt.xlabel("room_type")
plt.ylabel("minimum_nights")

plt.show()
```

###**We got the information that People are choosing Entire home,apartment or Private room which are located in Manhattan, Brooklyn, Queens**
